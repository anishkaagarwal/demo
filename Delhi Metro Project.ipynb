```python
import pandas as pd
import folium 


metro_data = pd.read_csv("Delhi-Metro-Network.csv")

print(metro_data.head())
```

```python
line_colors = {
    'Red line': 'red',
    'Blue line': 'blue',
    'Yellow line': 'beige',
    'Green line': 'green',
    'Voilet line': 'purple',
    'Pink line': 'pink',
    'Magenta line': 'darkred',
    'Orange line': 'orange',
    'Rapid Metro': 'cadetblue',
    'Aqua line': 'black',
    'Green line branch': 'lightgreen',
    'Blue line branch': 'lightblue',
    'Gray line': 'lightgray'
}

delhi_map_with_line_tooltip = folium.Map(location=[28.7042, 77.1025], zoom_start=10)

for index, row in metro_data.iterrows():
    line = row['Line']
    color = line_colors.get(line, 'black')  
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['Station Name']}",
        tooltip=f"{row['Station Name']}, {line}",
        icon=folium.Icon(color=color)
    ).add_to(delhi_map_with_line_tooltip)

delhi_map_with_line_tooltip
```

```python
import matplotlib.pyplot as plt

lines = [
    "Red Line",
    "Yellow Line",
    "Blue Line",
    "Green Line",
    "Violet Line",
    "Orange Line (Airport Express)",
    "Pink Line",
    "Magenta Line",
    "Grey Line",
    "Aqua Line",
    "Blue branch line",
    "Green branch line",
    "Rapid Metro"
]

stations = [29, 37, 50, 23, 34, 6, 38, 25, 3, 21,8,3,11]

plt.figure(figsize=(12, 6))
plt.bar(lines, stations, color='darkred')
plt.xlabel('Metro Lines')
plt.ylabel('Number of Stations')
plt.title('Number of Stations on Each Delhi Metro Line (2024)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

stations = ["Rajiv Chowk", "Kashmere Gate", "Central Secretariat", "Huda City Centre", "Noida City Centre",
    "Chandni Chowk", "New Delhi", "Anand Vihar", "Dwarka Sec-21", "Karol Bagh"]
station_ridership = [250, 220, 210, 200, 190, 180, 170, 160, 150, 140]

plt.figure(figsize=(12, 6))
plt.bar(stations, station_ridership, color='darkred')

plt.xlabel('Stations')
plt.ylabel('Ridership data(people in thousands)')
plt.title('Most crowded stations of Delhi Metro')
plt.xticks(rotation=60, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

hours = np.arange(5, 24)  
hourly_ridership = [5, 10, 20, 40, 120, 100, 180, 100, 50, 50, 60, 80, 90, 100, 120, 40, 20, 10, 5]

plt.figure(figsize=(12, 6))
plt.plot(hours, hourly_ridership, marker='o', linestyle='-', color='b')

plt.axvspan(8, 12, color='orange', alpha=0.3, label='Morning Peak Hours')
plt.axvspan(17, 20, color='red', alpha=0.3, label='Evening Peak Hours')

plt.xlabel('Hour of the Day')
plt.ylabel('Ridership (people in thousands)')
plt.title('Hourly Ridership of Delhi Metro, with Peak Hours')
plt.grid(True)
plt.xticks(hours)
plt.legend()

plt.show()
```
